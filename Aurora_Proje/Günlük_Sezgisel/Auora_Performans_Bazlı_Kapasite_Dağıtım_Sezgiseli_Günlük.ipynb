```python
df = pd.read_excel("Aurora.xlsx")
```

```python
df.info()
```

```python
df = df.rename(columns = {"Historik data": "Historik_Data", "Acente 1": "Acente_1", "Acente 2": "Acente_2", "Acente 3": "Acente_3","Acente 4": "Acente_4" })
```

```python
df
```

```python
df["Günlük_Toplam_Satış"] = df["Acente_1"] + df["Acente_2"]+ df["Acente_3"]+ df["Acente_4"]
```

```python
df
```

```python
df["Acente_1_Günlük_Skor"] = (df["Acente_1"]/df["Günlük_Toplam_Satış"]).round(2)
df["Acente_2_Günlük_Skor"] = (df["Acente_2"]/df["Günlük_Toplam_Satış"]).round(2)
df["Acente_3_Günlük_Skor"] = (df["Acente_3"]/df["Günlük_Toplam_Satış"]).round(2)
df["Acente_4_Günlük_Skor"] = (df["Acente_4"]/df["Günlük_Toplam_Satış"]).round(2)
```

```python
df
```

```python
gün_bilgisi =input("Kapasite dağılımını görmek istediğiniz gün bilgisini (Pazartesi, Salı, ..., Pazar) şeklinde giriniz: ")
# Select box olarak günlere yer verilebilir.
```

```python
filtre = df["Gün"] == gün_bilgisi
son_Acente_1_Günlük_Skor = df.loc[filtre, "Acente_1_Günlük_Skor"].iloc[-1]
son_Acente_2_Günlük_Skor = df.loc[filtre, "Acente_2_Günlük_Skor"].iloc[-1]
son_Acente_3_Günlük_Skor = df.loc[filtre, "Acente_3_Günlük_Skor"].iloc[-1]
son_Acente_4_Günlük_Skor = df.loc[filtre, "Acente_4_Günlük_Skor"].iloc[-1]
```

```python
print(son_Acente_1_Günlük_Skor)
print(son_Acente_2_Günlük_Skor)
print(son_Acente_3_Günlük_Skor)  
print(son_Acente_4_Günlük_Skor)  

# Bunu scripte koymaya gerek yok!
```

```python
df_grouped= df.groupby("Gün").sum(numeric_only=True).sort_values("Gün")
df_grouped= df_grouped.drop(columns =["Acente_1_Günlük_Skor", "Acente_2_Günlük_Skor", "Acente_3_Günlük_Skor","Acente_4_Günlük_Skor"])
df_grouped
```

```python
df_grouped["Günlük_Toplam_Satış"] = df_grouped["Acente_1"] + df_grouped["Acente_2"]+ df_grouped["Acente_3"]+ df_grouped["Acente_4"]
```

```python
df_grouped
```

```python
df_grouped["Acente_1_Historic_Skor"] = (df_grouped["Acente_1"] / df_grouped["Günlük_Toplam_Satış"]).round(2)
df_grouped["Acente_2_Historic_Skor"] = (df_grouped["Acente_2"] / df_grouped["Günlük_Toplam_Satış"]).round(2)
df_grouped["Acente_3_Historic_Skor"] = (df_grouped["Acente_3"] / df_grouped["Günlük_Toplam_Satış"]).round(2)
df_grouped["Acente_4_Historic_Skor"] = (df_grouped["Acente_4"] / df_grouped["Günlük_Toplam_Satış"]).round(2)
```

```python
df_grouped = df_grouped.reset_index() 
df_grouped
```

```python
beta =int(input("Kalan kapasitenin geçmiş performansa bağlı olarak dağıtılacak yüzdesini (0 ile 100 arasında)giriniz: "))
alpha =int(input("Kalan kapasitenin gün içi performansa bağlı olarak dağıtılacak yüzdesini (0 ile 100 arasında)giriniz: "))
kapasite =int(input("Acentelere dağıtalacak kontenjan miktarını giriniz: "))
```

```python
acente_1_kapasite_önerisi = int((((beta / 100 * kapasite * df_grouped[df_grouped["Gün"] == gün_bilgisi]["Acente_1_Historic_Skor"].iloc[-1]))+((1-beta /100) * kapasite * alpha / 100 * son_Acente_1_Günlük_Skor) + ((1-beta /100) * kapasite * (1 - alpha / 100) / 4 )).round(0))
acente_2_kapasite_önerisi = int((((beta / 100 * kapasite * df_grouped[df_grouped["Gün"] == gün_bilgisi]["Acente_2_Historic_Skor"].iloc[-1]))+((1-beta /100) * kapasite * alpha / 100 * son_Acente_2_Günlük_Skor) + ((1-beta /100) * kapasite * (1 - alpha / 100) / 4 )).round(0))
acente_3_kapasite_önerisi = int((((beta / 100 * kapasite * df_grouped[df_grouped["Gün"] == gün_bilgisi]["Acente_3_Historic_Skor"].iloc[-1]))+((1-beta /100) * kapasite * alpha / 100 * son_Acente_3_Günlük_Skor) + ((1-beta /100) * kapasite * (1 - alpha / 100) / 4 )).round(0))
acente_4_kapasite_önerisi =  kapasite - (acente_1_kapasite_önerisi + acente_2_kapasite_önerisi + acente_3_kapasite_önerisi)
```

```python
tarih = df[df["Gün"] == gün_bilgisi]["Historik_Data"].iloc[-1].date()
tarih = tarih + timedelta(days=7)
```

```python
print(f"Acente 1 için {tarih} tarihli {gün_bilgisi} günü kapasite önerisi: {acente_1_kapasite_önerisi}")
print(f"Acente 2 için {tarih} tarihli {gün_bilgisi} günü kapasite önerisi: {acente_2_kapasite_önerisi}")
print(f"Acente 3 için {tarih} tarihli {gün_bilgisi} günü kapasite önerisi: {acente_3_kapasite_önerisi}")
print(f"Acente 4 için {tarih} tarihli {gün_bilgisi} günü kapasite önerisi: {acente_4_kapasite_önerisi}")
```
